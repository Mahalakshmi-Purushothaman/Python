```python
# Python Libraries
# Scipy - It is used for scientific and technical computing. It is a colln of mathematical alg & convenience functions
# It is built on Numpy Extension. Installation: !pip install scipy
# scipy.cluster.vq - Vector Quantization - technique used for quantizing vectors. 

import numpy as np
from scipy.cluster.vq import kmeans, whiten
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_style("darkgrid")

#Create 50 datapoints in 2 clusters a and b 
pts = 100
a = np.random.multivariate_normal([0, 0], [[4, 1], [1, 4]], size = pts)
b = np.random.multivariate_normal([30, 10], [[10, 2], [2, 1]], size = pts)
features = np.concatenate((a, b))

# Whiten Data
Whitened = whiten(features)

# Find 2 clusters in the dataa
codebook, distortion = kmeans(Whitened, 2)

# Plot whitened data and cluster centers in red
plt.scatter(Whitened[:, 0], Whitened[:, 1])
plt.scatter(codebook[:, 0], codebook[:, 1], c='r')
plt.show()
```

```python
# Scipy - Interpolation
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 10)
y = np.cos(x**2/3+4)

plt.scatter(x, y, c='r')
plt.show()
```

```python
# Scipy - Interpolation
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

fun1 = interp1d(x,y,kind = 'linear')
fun2 = interp1d(x,y,kind = 'cubic')

xnew=np.linspace(0,4,30)

plt.plot(x, y, 'o', xnew, fun1(xnew), '-', xnew, fun2(xnew), '--')
plt.legend(['data', 'linear', 'cubic', 'nearest'], loc = 'best')
plt.show()
```

```python
# we are trying to solve a linear algebra system which can be given as 
# x+3y-3z = -3 
# 2x-5y+4z = 13
# 5x+4y-z = 5 
# find values of x, y, z for which all these eqn are zero

import numpy as np
from scipy import linalg

# Input array 
a = np.array([[1,2,-3], [2,-5,4], [5,4,-1]])

# Solution array 
b = np.array([[-3], [13], [5]])

#Solve the linear algebra 
x=linalg.solve(a,b)
print(x)

#Check results
print('\n Checking results must be zeros')
print(a.dot(x) -b)
```
